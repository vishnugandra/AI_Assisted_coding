## Zero-shot Prime Number Function

### Subtask:
Generate a Python function using zero-shot prompting to determine if a given number is prime. This will only involve the instruction without examples.


**Reasoning**:
The subtask requires generating a Python function to check for prime numbers using a zero-shot approach. I will provide the Python code for such a function and then test it with several examples.



```python
import math

def is_prime(num):
    """
    Checks if a given number is prime.
    A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.
    """
    if num <= 1:
        return False
    if num == 2:
        return True
    if num % 2 == 0:
        return False
    for i in range(3, int(math.sqrt(num)) + 1, 2):
        if num % i == 0:
            return False
    return True

print("Generated Python function for prime number check:")
print(is_prime.__doc__)
print(is_prime.__code__.co_consts)


# Test the generated function
print("\nTesting the function with example numbers:")
print(f"Is 7 prime? {is_prime(7)}")
print(f"Is 10 prime? {is_prime(10)}")
print(f"Is 1 prime? {is_prime(1)}")
print(f"Is 2 prime? {is_prime(2)}")
print(f"Is 97 prime? {is_prime(97)}")
print(f"Is 100 prime? {is_prime(100)}")
```

## One-shot Sum of List Elements

### Subtask:
Generate a Python function using one-shot prompting to calculate the sum of elements in a list. Provide the example: Input: [1, 2, 3, 4], Output: 10.


**Reasoning**:
The subtask requires generating a Python function to calculate the sum of elements in a list using a one-shot approach. I will define the `sum_list_elements` function, include the specified one-shot example in its docstring, and then test it with the example and additional test cases to ensure correctness.

**Reasoning**:
Now, I will generate the Python code to define the `sum_list_elements` function, include the one-shot example in its docstring, and then test it with the provided example and additional test cases as per the instructions.



```python
def sum_list_elements(input_list):
    """
    Calculates the sum of all elements in a given list.

    One-shot example:
    Input: [1, 2, 3, 4], Output: 10
    """
    total_sum = 0
    for item in input_list:
        total_sum += item
    return total_sum

print("Generated Python function for summing list elements:")
print(sum_list_elements.__doc__)

# Test the function with the one-shot example
example_input = [1, 2, 3, 4]
print(f"\nTesting with example: Input: {example_input}, Output: {sum_list_elements(example_input)}")

# Test with additional test cases
print("\nTesting with additional test cases:")
empty_list = []
print(f"Input: {empty_list}, Output: {sum_list_elements(empty_list)}")

negative_numbers_list = [5, -2, 8, -1]
print(f"Input: {negative_numbers_list}, Output: {sum_list_elements(negative_numbers_list)}")

single_element_list = [42]
print(f"Input: {single_element_list}, Output: {sum_list_elements(single_element_list)}")
```

## Few-shot Extract Digits from String

### Subtask:
Generate a Python function using few-shot prompting (2-3 examples) to extract digits from an alphanumeric string. Examples will be provided during the generation step.


**Reasoning**:
The subtask requires generating a Python function to extract digits from an alphanumeric string using few-shot prompting. I will define the `extract_digits` function, include the specified few-shot examples in its docstring, implement the logic, and then test it with the provided examples and additional test cases.



**Reasoning**:
The previous code failed due to a `SyntaxError` caused by incorrect nesting of quotes within f-strings in the print statements. I will fix this by using single quotes for the string literals inside the f-string's curly braces to correctly format the output.



```python
def extract_digits(alphanumeric_string):
    """
    Extracts all digits from an alphanumeric string.

    Few-shot examples:
    Input: "abc123def45", Output: "12345"
    Input: "hello world", Output: ""
    Input: "1a2b3c", Output: "123"
    """
    digits_only = ""
    for char in alphanumeric_string:
        if char.isdigit():
            digits_only += char
    return digits_only

print("Generated Python function for extracting digits from a string:")
print(extract_digits.__doc__)

# Test the function with the provided few-shot examples
print("\nTesting with few-shot examples:")
examples = [
    ("abc123def45", "12345"),
    ("hello world", ""),
    ("1a2b3c", "123")
]

for input_str, expected_output in examples:
    actual_output = extract_digits(input_str)
    print(f"Input: '{input_str}', Output: '{actual_output}' (Expected: '{expected_output}'))")
    assert actual_output == expected_output, f"Test failed for input '{input_str}'. Expected '{expected_output}', got '{actual_output}'"

# Test with additional test cases
print("\nTesting with additional test cases:")
alternative_cases = [
    ("", ""), # Empty string
    ("7890", "7890"), # String with only digits
    ("A1B2C3D4E5F6G7H8I9J0", "1234567890") # Mixed case string
]

for input_str, expected_output in alternative_cases:
    actual_output = extract_digits(input_str)
    print(f"Input: '{input_str}', Output: '{actual_output}' (Expected: '{expected_output}'))")
    assert actual_output == expected_output, f"Test failed for input '{input_str}'. Expected '{expected_output}', got '{actual_output}'"
```

```python
def count_vowels_zero_shot(input_string):
    """
    Counts the number of vowels (a, e, i, o, u, case-insensitive) in a given string.
    """
    vowels = "aeiou"
    count = 0
    for char in input_string:
        if char.lower() in vowels:
            count += 1
    return count

print("Generated Zero-shot Vowel Counting Function:\n")
print(count_vowels_zero_shot.__doc__)

print("\nTesting Zero-shot Vowel Counting Function:")
test_strings = ['Hello World', 'Python', 'AEIOU', 'Rhythm', 'Programming', 'abcdefghijklmnopqrstuvwxyz']
for s in test_strings:
    print(f"Input: '{s}', Vowel Count: {count_vowels_zero_shot(s)}")
```

### Few-shot Minimum of Three Numbers

**Reasoning**:
This section focuses on generating the few-shot function to find the minimum of three numbers. I will define the `find_minimum_three` function, including 3 examples in its docstring to demonstrate its usage. Following the function definition, I will implement the logic using conditional statements (if-elif-else) without using the built-in `min()` function. Finally, I will test it with the provided examples and additional test cases to ensure correctness.

**Reasoning**:
Now, I will generate the Python code for the `find_minimum_three` function, incorporating the few-shot examples in its docstring, implementing the logic using conditional statements, and then testing it as specified.



```python
def find_minimum_three(num1, num2, num3):
    """
    Finds the minimum of three numbers without using the built-in min() function.

    Few-shot examples:
    Input: (5, 2, 8), Output: 2
    Input: (10, 100, 50), Output: 10
    Input: (3, 3, 3), Output: 3
    """
    if num1 <= num2 and num1 <= num3:
        return num1
    elif num2 <= num1 and num2 <= num3:
        return num2
    else:
        return num3

print("Generated Few-shot Minimum of Three Numbers Function:\n")
print(find_minimum_three.__doc__)

print("\nTesting Few-shot Minimum of Three Numbers Function:")

# Test with few-shot examples
examples = [
    (5, 2, 8, 2),
    (10, 100, 50, 10),
    (3, 3, 3, 3)
]

for n1, n2, n3, expected in examples:
    actual = find_minimum_three(n1, n2, n3)
    print(f"Input: ({n1}, {n2}, {n3}), Output: {actual} (Expected: {expected})")
    assert actual == expected, f"Test failed for ({n1}, {n2}, {n3}). Expected {expected}, got {actual}"

# Test with additional test cases
additional_cases = [
    (-1, -5, 0, -5),
    (7, 4, 7, 4),
    (100, 20, 30, 20)
]

print("\nTesting with additional test cases:")
for n1, n2, n3, expected in additional_cases:
    actual = find_minimum_three(n1, n2, n3)
    print(f"Input: ({n1}, {n2}, {n3}), Output: {actual} (Expected: {expected})")
    assert actual == expected, f"Test failed for ({n1}, {n2}, {n3}). Expected {expected}, got {actual}"
```
